```python
import numpy as np
import random
import torch
from src.data import load_digits_data
from src.training import train_multiseed
from src.plots import plot_ex2_multiseed

torch.cuda.empty_cache()
device = "cuda" if torch.cuda.is_available() else "cpu"

data_seed = 0
run_seeds = list(range(5))
```

```python
epochs = 5000
n = 10
m = 30000
clean_beta = n*1e05
noisy_beta = n*1e03
```

```python
torch.manual_seed(0)
np.random.seed(0)
random.seed(0)

data = load_digits_data(n=n, random_labels=True, device=device, seed=data_seed)
d = data["X_train"].shape[1]

common = dict(
    seeds=run_seeds,
    data=data,
    eta=1e-5,
    epochs=epochs,
    m=m,
    init_type="standard",
    regularization_scale=1.0,
    track_jacobian=True, 
    device=device,
    print_every=epochs//10,
)

results = {
    "clean": train_multiseed(beta=clean_beta, use_linearized=False, **common,),
    "noisy": train_multiseed(beta=noisy_beta, use_linearized=True, **common,),
}
```

```python
plot_ex2_multiseed(results)
```
